```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import svm
```

```python
data  = pd.read_csv("gender.csv")
data.head()
```

```python
data['Class'] = data['Gender'].apply(lambda x: 'r' if x=='F' else 'b')
```

```python
data.head()
```

```python
data = data.iloc[:25]
```

```python
fig ,ax =plt.subplots()
ax.scatter(x=data['Height'],y=data['Hair length'],c=data['Class'])
ax.axhline(19)
x = np.array([150,190])
ax.plot(x,x-147,c='y')
ax.plot(x,2*x-315,c='g')
```

* which line is the most accurate

```python
X = data[['Height','Hair length']]
y = data['Gender']
y = np.array([1 if i=='M' else 0 for i in y])
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X,y)
```

```python
X_test = np.random.rand(10000,2)
X_test = X_test*(50,150) + (150,0)
y_preds = clf.predict(X_test)
```

```python
fig,ax = plt.subplots()
ax.scatter(x=X_test[:,0],y=X_test[:,1],c=['y' if i==1 else 'g' for i in y_preds],alpha=0.25)
ax.scatter(x=X['Height'],y=X['Hair length'],c=['r' if i==1 else 'b' for i in y])
```
